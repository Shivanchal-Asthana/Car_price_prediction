```python
#creating empty list for urls
product_urls=[]
urls=[]
#1000 times we scroll the page by getting more details
start=0
end=1000
for i in range(start,end):
    time.sleep(2)
    driver.execute_script("window.scrollBy(0,document.body.scrollHeight)") 

#scraping urls
url=driver.find_elements_by_xpath('//a[@class="_9Ue0B"]')
for i in url:
    product_urls.append(i.get_attribute("href"))

#creating empty list
car_name2 = []
price2=[]
km_driven2 =[]
location2 =[]
brand_name2=[]
owner2=[]
emi2=[]
fuel2=[]
gear_type2=[]

#fetching the details of 'New Delhi' location
for url in product_urls:
    driver.get(url)
    time.sleep(2)
    
    #price
    try:
        p = driver.find_elements_by_xpath('//*[@class="_7udZZ"]//span')
        for i in p:
            price2.append(i.text)
    except NoSuchElementException:
        price2.append('-')
        
    #brand name   
    try:
        p = driver.find_elements_by_xpath('//*[@class="_3FpCg"]')
        for i in p:
            brand_name2.append(i.text)
    except NoSuchElementException:
        brand_name2.append('-')
        
    #car name   
    try:
        p = driver.find_elements_by_xpath('//*[@class="cvakB"]')
        for i in p:
            car_name2.append(i.text)
    except NoSuchElementException:
        car_name2.append('-')
        
    #gear type    
    try:
        p = driver.find_elements_by_xpath('//*[@class="cvakB"]//span')
        for i in p:
            gear_type2.append(i.text)
    except NoSuchElementException:
        gear_type2.append('-')
        
    #KM Driven
    try:
        p = driver.find_elements_by_xpath('//*[@class="bVR0c"]//li[1]')
        for i in p:
            km_driven2.append(i.text)
    except NoSuchElementException:
        km_driven2.append('-')
        
    #Owner
    try:
        p = driver.find_elements_by_xpath('//*[@class="bVR0c"]//li[2]')
        for i in p:
            owner2.append(i.text)
    except NoSuchElementException:
        owner2.append('-')
        
    #Fuel
    try:
        p = driver.find_elements_by_xpath('//*[@class="bVR0c"]//li[3]')
        for i in p:
            fuel2.append(i.text)
    except NoSuchElementException:
        fuel2.append('-')
        
    #EMI
    try:
        p = driver.find_elements_by_xpath('//*[@class="_2HFRN"]//strong')
        for i in p:
            emi2.append(i.text)
    except NoSuchElementException:
        emi2.append('-')
        
    #Location
    try:
        p = driver.find_elements_by_xpath('//*[@class="_7udZZ"]//span')
        for i in p:
            location2.append("New Delhi")
    except NoSuchElementException:
        location2.append('-')
```
